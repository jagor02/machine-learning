## Import potrzebnych bibliotek

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Zadanie - baza danych Iris

```python
# załadowanie bazy danych
from sklearn.datasets import load_iris
irisRaw = load_iris()

# konwersja danych do data frame'u z pakietu 'pandas' wraz z nazwami kolumn
iris = pd.DataFrame(data= np.c_[irisRaw['data'], irisRaw['target']],
                     columns= irisRaw['feature_names'] + ['target'])
```

```python
iris_groupby_by_df.var() # wariancja wartości poszczególnych zmiennych (kolumny) dla podziału względem 'target'
```

```python
iris_groupby_by_df.std() # odchylenie standardowe wartości poszczególnych zmiennych (kolumny) dla podziału względem 'target'
```

## 10. Wyodrębnij wiersze dla których długość kielicha kwiatów (sepal length) jest większa od średniej długości (cały zbiór).

```python
iris.loc[iris['sepal length (cm)'] > iris['sepal length (cm)'].mean()]
```

```python
vars = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)'] # nazwy zmiennych
colors = ['green', 'blue', 'purple'] # kolory wykresów

for i in range(4): # po wszystkich zmiennych
  # wstępne ustawienia wykresów
  fig, (ax0, ax1, ax2) = plt.subplots(nrows=1, ncols=3)
  fig.set_figheight(6)
  fig.set_figwidth(18)
  fig.suptitle(''.join(['Histogramy dla zmiennej ', vars[i]]), fontsize=22) # główny tytuł wykresów
  # wyliczanie 'wysokości' poszczególnych prostokątów i współrzędnych x-owych ich boków dla badanego histogramu; KLASA 0
  counts, bins = np.histogram(iris.loc[iris['target'] == 0, vars[i]])
  ax0.hist(bins[:-1], bins, weights=counts, facecolor=colors[0]) # rysowanie histogramu
  ax0.set_xlabel('Długość (cm)')
  ax0.set_ylabel('Ilość danych irysów')
  ax0.set_title('Klasa irysów - 0')
  # wyliczanie 'wysokości' poszczególnych prostokątów i współrzędnych x-owych ich boków dla badanego histogramu; KLASA 1
  counts, bins = np.histogram((iris[vars[i]]).loc[iris['target'] == 1])
  ax1.hist(bins[:-1], bins, weights=counts, facecolor=colors[1]) # rysowanie histogramu
  ax1.set_xlabel('Długość (cm)')
  ax1.set_ylabel('Ilość danych irysów')
  ax1.set_title('Klasa irysów - 1')
  # wyliczanie 'wysokości' poszczególnych prostokątów i współrzędnych x-owych ich boków dla badanego histogramu; KLASA 2
  counts, bins = np.histogram((iris[vars[i]]).loc[iris['target'] == 2])
  ax2.hist(bins[:-1], bins, weights=counts, facecolor=colors[2]) # rysowanie histogramu
  ax2.set_xlabel('Długość (cm)')
  ax2.set_ylabel('Ilość danych irysów')
  ax2.set_title('Klasa irysów - 2')
  plt.show()
```
